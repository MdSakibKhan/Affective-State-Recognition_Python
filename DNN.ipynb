```
import pandas as pd
dataset = pd.read_csv('/content/modifiedData.csv')
dataset = pd.DataFrame(dataset)
data = dataset
data
```

```
#Arousal
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2

classifier = Sequential() #ANN

classifier.add(Dense(units=42,activation='relu',input_dim=630))
classifier.add(Dense(units=14,activation='sigmoid'))
classifier.add(Dense(units=7,activation='sigmoid'))
classifier.add(Dense(units=1,activation='sigmoid'))


classifier.compile(optimizer='rmsprop', metrics=['accuracy'], loss='binary_crossentropy')


Y = data['Dominance']
X = data.drop(['Arousal','Valence','Dominance'], axis=1) 
X  = preprocessing.minmax_scale(X)

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.30, random_state=0, stratify = Y)

classifier.fit(X_train, Y_train, batch_size=1, epochs=100)

Y_pred = classifier.predict(X_test)
Y_pred = [1 if y>=0.5 else 0 for y in Y_pred]

from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```
